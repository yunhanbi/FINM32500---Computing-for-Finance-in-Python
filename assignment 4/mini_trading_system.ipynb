## Part 1: FIX Message Parser

- Role of FIX Parser:
    - FIX is the language of markets (Used in real e-trading, on exchanges, brokers, banks, market makers, hedge funds, etc.)
    - The message however, is just a string, and the trading system cannot operate on raw strings

-  What the Parser does:
    - Converts raw text into structured data
    - Validates required fields
    - everything downstream depends on it

- "The FIX parser converts raw exchange messages into validated structured data so the trading system can safely process orders."
